### Ödev 1

liste = ["Python",True,9,"3",8.4,"Hi-Kod","False",4.7]

1. Aşağıdaki işlemleri indexing ve slicing kullanarak liste üzerinde uygulayın.

   - "3" değerine ulaşmak için indexleme yapın.
   - "Hi-Kod" değerine ulaşmak için indexleme yapın.
   - 4.7 değerine ulaşmak için indexleme yapın.
   - 9,"3",8.4,"Hi-Kod" değerlerine ulaşmak için slicing yapın.
   - 8.4,"Hi-Kod","False",4.7 değerlerine ulaşmak için slicing yapın.

```python
liste = ["Python",True,9,"3",8.4,"Hi-Kod","False",4.7]
print(liste[3])
print(liste[5])
print(liste[7])
print(liste[2:6])
print(liste[4:])
```

### Ödev 2

Verilen listede bulunan string veri tipindeki öğeleri yeni_liste isimli listeye eklenir.

liste = ["Python",True,9,"3",8.4,"Hi-Kod","False",4.7]

```python
liste = ["Python",True,9,"3",8.4,"Hi-Kod","False",4.7]
yeni_liste = []

for i in liste:
    if type(i) == str:
        yeni_liste.append(i)
print(yeni_liste)
```

## Sozluk

1. Bir sözlük oluşturulur ve bu sözlükte öğrencilerin isimleri ve Matematik, Fizik, Kimya notları tutulur. Kullanıcıdan isim ve ders ismi(Matematik, Fizik, Kimya) istenir ve bu bilgilere göre çıktı verilir.

2. Sözlük üzerinde değerleri değiştirme, yeni değer ekleme, kullanıcıya ulaşmak istediği bilgileri sorma gibi uygulamalar yapın.

```python
notlar = {
    "Serife": {"Matematik": 95, "Fizik": 85, "Kimya": 90},
    "Beyza": {"Matematik": 100, "Fizik": 50, "Kimya": 45},
    "Eylul": {"Matematik": 90, "Fizik": 60, "Kimya": 65}
}

isim = input("Bilgilerini görmek istediginiz ogrencinin adini giriniz=")
ders = input("Ogrenmek istediginiz dersin adini giriniz=")

if isim in notlar:
    if ders in notlar[isim]:
        print(f"{isim} ogrencisinin {ders} notu: {notlar[isim][ders]}")
    else:
        print("Girdiginiz ders bulunamadi")
else:
    print("Girdiginiz isimde ogrenci bulunamadi")
```

```python
# Başlangıç sözlüğü
notlar = {
    "Şerife": {"Matematik": 95, "Fizik": 85, "Kimya": 90},
    "Beyza": {"Matematik": 100, "Fizik": 50, "Kimya": 45},
    "Eylül": {"Matematik": 90, "Fizik": 60, "Kimya": 65}
}

print("--- Öğrenci Not Yönetim Sistemi ---")

while True:
    print("\nLütfen yapmak istediğiniz işlemi seçiniz:")
    print("1: Not Değiştirme (Var olan bir öğrencinin ders notunu güncelle)")
    print("2: Yeni Öğrenci Ekleme (Yeni öğrenci ve notlarını ekle)")
    print("3: Not Sorgulama (Belirli bir öğrencinin ders notunu gör)")
    print("4: Tüm Notları Göster")
    print("5: Çıkış")

    secim = input("Seçiminiz (1-5): ")

    if secim == '1':
        # 1. Değer Değiştirme
        isim = input("Notunu değiştirmek istediğiniz öğrencinin adını girin: ")
        if isim in notlar:
            ders = input("Değiştirmek istediğiniz dersin adını girin (Matematik/Fizik/Kimya): ").capitalize()
            if ders in notlar[isim]:
                while True:
                    try:
                        yeni_not = int(input(f"{isim} öğrencisinin {ders} dersi için yeni notu girin: "))
                        if 0 <= yeni_not <= 100:
                            notlar[isim][ders] = yeni_not  # Sözlükteki var olan değeri değiştirme
                            print(f"Başarılı. {isim} öğrencisinin {ders} notu {yeni_not} olarak güncellendi.")
                            break
                        else:
                            print("Hata: Not 0 ile 100 arasında olmalıdır.")
                    except ValueError:
                        print("Hata: Geçersiz not girişi. Lütfen sayısal bir değer girin.")
            else:
                print(f"Hata: {isim} öğrencisinde '{ders}' dersi bulunamadı.")
        else:
            print(f"Hata: '{isim}' isimli öğrenci bulunamadı.")

    elif secim == '2':
        # 2. Yeni Değer Ekleme
        yeni_isim = input("Eklemek istediğiniz yeni öğrencinin adını girin: ")
        if yeni_isim not in notlar:
            try:
                mat = int(input(f"{yeni_isim} için Matematik notunu girin: "))
                fizik = int(input(f"{yeni_isim} için Fizik notunu girin: "))
                kimya = int(input(f"{yeni_isim} için Kimya notunu girin: "))

                # Yeni bir anahtar-değer çifti ekleme
                notlar[yeni_isim] = {"Matematik": mat, "Fizik": fizik, "Kimya": kimya}
                print(f"Başarılı. '{yeni_isim}' isimli öğrenci notlarıyla birlikte eklendi.")
            except ValueError:
                print("Hata: Notlar sayı olmalıdır. Ekleme işlemi başarısız.")
        else:
            print(f"Hata: '{yeni_isim}' isimli öğrenci zaten sistemde mevcut.")

    elif secim == '3':
        # 3. Bilgiye Ulaşma (Sorgulama)
        isim = input("Bilgileri görmek istediğiniz öğrencinin adını giriniz: ")
        ders = input("Öğrenmek istediğiniz dersin adını giriniz: ").capitalize()

        if isim in notlar:
            if ders in notlar[isim]:
                print(f"{isim} öğrencisinin {ders} notu: {notlar[isim][ders]}")
            else:
                print(f"Hata: Girdiğiniz ders ({ders}) bulunamadı.")
        else:
            print(f"Hata: Girdiğiniz isimde öğrenci ({isim}) bulunamadı.")

    elif secim == '4':
        # Tüm Veriyi Görüntüleme
        print("\n--- Mevcut Tüm Öğrenci Notları ---")
        if not notlar:
            print("Sistemde kayıtlı öğrenci bulunmamaktadır.")
        else:
            for ogrenci, not_bilgileri in notlar.items():
                print(f"Öğrenci: {ogrenci}")
                for ders, notu in not_bilgileri.items():
                    print(f"  - {ders}: {notu}")
        print("---------------------------------")


    elif secim == '5':
        # Çıkış
        print("Programdan çıkılıyor.")
        break

    else:
        print("Geçersiz seçim. Lütfen 1 ile 5 arasında bir sayı girin.")
```
